```python
# ML Final Project Visualizations
### Dataset source: Melbourne, Seattle, and Sydney

# Description: 
### To show a frequency diagram for the given attributes. Using bar charts, data is plotted.

import matplotlib.pyplot as plt
import mysklearn.myutils as myutils
import mysklearn.plot_utils as plot_utils
from mysklearn.mypytable import MyPyTable
import os
import copy


data_name = os.path.join("input_data", "MelbourneparsedData.csv")
table = MyPyTable().load_from_file(data_name)
table_copy = copy.deepcopy(table)
table_copy.remove_rows_with_missing_values()
MyPyTable.save_to_file(table_copy, data_name)

column = table_copy.get_column("RainToday")
length = len(column)
yes_count = 0
for col in column:
    if col == "Yes":
        yes_count = yes_count + 1
no_count = length - yes_count

fig2 = plt.figure()
plt.title("Days of Rain vs Days of No Rain in Melbourne")
plt.xlabel("Result")
plt.ylabel("Frequency of Result")
fig2.text(0, -.08, "Figure 1 shows a bar graph for the number of days with rain and days without rain in Melbourne")
plot_utils.bar_graph(['Yes', 'No'], [yes_count, no_count])
plt.show()

###parse rainToday's that are yes
###month values = month name (12 of them), month_counts = (counts of rainToday that are yes for each month)

### table with only yes' and no NA's
months, rain_col = myutils.get_aus_frequencies(table_copy, "RainToday")

## Platforms bar graph is being shown
fig2 = plt.figure()
plt.title("Total Number of Rainy Days per Month in Melbourne")
plt.xlabel("Months")
plt.ylabel("Frequency of Rainy Days")
fig2.text(0, -.08, "Figure 2 shows a bar graph for the total number of rainy days per month in Melbourne")
plot_utils.bar_graph(months, rain_col)
plt.figure()

fig2 = plt.figure()
plt.title("Percentage of Rainy Days in Melbourne")
fig2.text(0, -.08, "Figure 3 shows a pie chart for the percentages of rainy days per month distributed throughout Melbourne")
plot_utils.pie_chart(months, rain_col)
plt.figure()

### Melbourne EDA: The first data visualization is meant to show the distribution of days rained and days not rained in Melbourne. The second bar graph is meant to show which months rain the most and in seeing the trends, one can also get a better idea of the season/climate in Melbourne. The pie chart is meant to show the distribution of these 'days rained' per month in proportion to the whole dataset. This shows us that most every month apart from March-December rains about the same (climate doesn't change).
```
