```python

# Naive Bayes Confusion

import os
####-----------------------------------------------------------
### Melbourne
####-----------------------------------------------------------

mel_data.remove_rows_with_missing_values()
# Date, MinTemp, MaxTemp, Sunshine, predicting RainToday

date = mel_data.get_column('Date')
minTemp = mel_data.get_column('MinTemp')
maxTemp = mel_data.get_column('MaxTemp')
sun = mel_data.get_column('Sunshine')

X_train = [[date[i],minTemp[i],maxTemp[i],sun[i]] for i in range(len(date))]
y_train = mel_data.get_column('RainToday')

myb = MyNaiveBayesClassifier()
myb.fit(X_train, y_train)

strattrain_folds, strattest_folds = myevaluation.stratified_kfold_cross_validation(X_train, y_train, 10)
X_train_strat, y_train_strat, X_test_strat, y_test_strat = myutils.get_from_folds(X_train, y_train, strattrain_folds, strattest_folds)

predicted_naive = myb.predict(X_test_strat)
accuracy = 0
predicted_correct = 0
for i in range(len(y_test_strat)):
    if y_test_strat[i] == predicted_naive[i]:
        predicted_correct+=1
accuracy = predicted_correct/len(y_test_strat)

####-----------------------------------------------------------
### Sydney
####-----------------------------------------------------------

syd_data.remove_rows_with_missing_values()
# Date, MinTemp, MaxTemp, Sunshine, predicting RainToday

date2 = syd_data.get_column('Date')
minTemp2 = syd_data.get_column('MinTemp')
maxTemp2 = syd_data.get_column('MaxTemp')
sun2 = syd_data.get_column('Sunshine')

X_train2 = [[date2[i],minTemp2[i],maxTemp2[i],sun2[i]] for i in range(len(date2))]
y_train2 = syd_data.get_column('RainToday')

myb2 = MyNaiveBayesClassifier()
myb2.fit(X_train2, y_train2)

strattrain_folds2, strattest_folds2 = myevaluation.stratified_kfold_cross_validation(X_train2, y_train2, 10)
X_train_strat2, y_train_strat2, X_test_strat2, y_test_strat2 = myutils.get_from_folds(X_train2, y_train2, strattrain_folds2, strattest_folds2)

predicted_naive2 = myb2.predict(X_test_strat2)
accuracy2 = 0
predicted_correct2 = 0
for i in range(len(y_test_strat2)):
    if y_test_strat2[i] == predicted_naive2[i]:
        predicted_correct2+=1
accuracy2 = predicted_correct2/len(y_test_strat2)

####-----------------------------------------------------------
### Seattle
####-----------------------------------------------------------

sea_data.remove_rows_with_missing_values()
# DATE,PRCP,TMAX,TMIN,RAIN

date3 = sea_data.get_column('DATE')
minTemp3 = sea_data.get_column('TMIN')
maxTemp3 = sea_data.get_column('TMAX')

X_train3 = [[date3[i],maxTemp3[i],minTemp3[i]] for i in range(len(date3))]
y_train3 = sea_data.get_column('RAIN')

myb3 = MyNaiveBayesClassifier()
myb3.fit(X_train3, y_train3)

strattrain_folds3, strattest_folds3 = myevaluation.stratified_kfold_cross_validation(X_train3, y_train3, 10)
X_train_strat3, y_train_strat3, X_test_strat3, y_test_strat3 = myutils.get_from_folds(X_train3, y_train3, strattrain_folds3, strattest_folds3)

predicted_naive3 = myb3.predict(X_test_strat3)
accuracy3 = 0
predicted_correct3 = 0
for i in range(len(y_test_strat3)):
    if y_test_strat3[i] == predicted_naive3[i]:
        predicted_correct3+=1
accuracy3 = predicted_correct3/len(y_test_strat3)

print('\n===========================================\n\
MyNaiveBayesClassifier Classifier Results\n\
===========================================\n')
print('Melbourne accuracy: ', accuracy)

print('Sydney accuracy: ', accuracy2)

print('Seattle accuracy: ', accuracy3)

# Notes
## Evaluated with stratified k fold cross validation correct folds to predict on with Naive Bayes
```
